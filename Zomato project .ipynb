```python
#numbers of rating and rate out of 5 
dataset.groupby('rate (out of 5)')['num of ratings'].count().sort_values(ascending=True).head(10).plot(kind='pie',autopct='%1.2f%%')
```

```python
#total online orders by cuisines type
dataset[dataset['online_order']=='Yes'].groupby('online_order')['cuisines type'].count().plot(kind='bar')
```

```python
#number of restaurants by location
restaurant_by_area = dataset['area'].value_counts()
restaurant_by_area.plot(kind='bar')
```

```python
#Average rating by location
ave_rating_by_location = dataset.groupby('area')['rate (out of 5)'].mean().sort_values(ascending = False)
ave_rating_by_location.plot(kind='bar')
```

```python
#count(Distribution) of the restaurant type
restaurant_type_counts= dataset['restaurant type'].value_counts().head(10)
restaurant_type_counts.plot(kind='bar')
```

```python
#online orders
online_order = dataset['online_order'].value_counts()
online_order.plot(kind='bar')
```

```python
#tables bookings
table_bookings = dataset['table booking'].value_counts()
table_bookings.plot(kind='bar')
```
